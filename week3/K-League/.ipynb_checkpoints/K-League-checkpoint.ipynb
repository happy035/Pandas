```python
p_names = pd.read_json('players.csv')
p_names.set_index('p_id',inplace=True)
p_names.head()

#p_id가 인덱스이므로 ix 속성으로 선택
p_names.ix[19970001]
```

```python
#팀으로 그루핑
gb_team = k.groupby('team')

#그룹의 key 확인
gb_team.groups.keys()
```

```python
team1 = gb_team.get_group('부산').sort_values('date')
team1.head()
```

```python
def win(x):
    if x['goal'] > x['lose']:
        return 1
    else: return 0

def lost(x):
    if x['goal'] < x['lose']:
        return 1
    else: return 0 

def draw(x):
    if x['goal'] == x['lose']:
        return 1
    else: return 0

# data별로 그룹핑

gb_team1_date = team1.groupby('date')

res_df = gb_team1_date.agg({'goal' : sum, 'lose' : sum, 'self_goal' : sum})

# 계산하기 쉽게 승(win), 패(lost), 무(draw)를 정수로 저장
res_df['win'] = res_df.apply(win, axis=1)
res_df['lost'] = res_df.apply(lost, axis=1)
res_df['draw'] = res_df.apply(draw, axis=1)

res_df
```

```python
print( res_df['win'].sum(),
      res_df['draw'].sum(),
      res_df['lost'].sum(),
      res_df['goal'].sum() + res_df['self_goal'].sum(),
      res_df['lose'].sum(),
      res_df['win'].sum() * 3 + res_df['draw'].sum())
```

```python
def team_result(k, team_name):
    gb = k.groupby('team')
    team = gb.get_group(team_name).sort_values('date')
    gb_date = team.groupby('date')
    res = gb_date.agg({'goal' : sum, 'lose' : sum, 'self_goal' : sum})
    res['win'] = res.apply(win, axis=1)
    res['lost'] = res.apply(lost, axis=1)
    res['draw'] = res.apply(draw, axis=1)
    
    # 파이썬 사전으로 데이터 생성
    d = {
        'Team' : team_name,
        'Match' : len(gb_date.groups),
        'Win' : res['win'].sum(),
        'Draw' : res['draw'].sum(),
        'Lose' : res['lost'].sum(),
        'GF' : res['goal'].sum() + res['self_goal'].sum(),
        'GA' : res['lose'].sum()
    }
    
    #골 득실과 승점도 계산, 득점(GF) - 실점(GA)
    d['GD'] = d['GF'] - d['GA']
    d['pts'] = d['Win'] * 3 + d['Draw']
    
    return d

team_result(k,'부산')
```

```python
table = pd.DataFrame([team_result(k, team_name) for team_name in gb_team.groups.keys()])

# 재 인덱싱
table = table.reindex_axis(['Team','Match','Win', 'Draw','Lose','GF','GA','GD', 'pts'], axis=1)
```

```python
table.sort_values('pts', ascending=False)
```
