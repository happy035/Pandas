```python
'''
FTHG = Full Time Home Team Goal
FTAG = Full Time Away Team Goal 
FTR = Full Time Result (H / A / D)
'''

working_df = df[['Date', 'HomeTeam', 'AwayTeam','FTHG','FTAG','FTR']]


#GF = Goal For 득점
#GA = Goal Again 실점
#GD = Goal difference 차
#Point = W*3 + D

teams = df['HomeTeam'].unique()
teams
```

```python
full_table = pd.DataFrame({'Team' : teams}, columns={'Team','Win','Draw','Lost','GF','GA','GD','Point'})
full_table = full_table.fillna(0)
full_table = full_table.reindex_axis(['Team','Win','Draw','Lost','GF','GA','GD','Point'], axis=1)
full_table.head()
```

```python
working_df.head()
```

```python
#HomeTeam으로 그룹핑
home_gb = working_df.groupby('HomeTeam')
'''
for name, group in home_gb:
    print(name)
    print(group)
'''
d = dict(list(home_gb))
```

```python
full_table.loc[full_table['Team'] == 'Arsenal','GF'] = d['Arsenal']['FTHG'].sum()
full_table.loc[full_table['Team'] == 'Arsenal','GA'] = d['Arsenal']['FTAG'].sum()
full_table['GD'] = full_table['GF'] - full_table['GA']

full_table
```

```python
home_gb.first()
home_gb.last()

def goal(group):
    gf = group['FTHG'].sum()
    ga = group['FTAG'].sum()
    gd = gf - ga

home_gb.agg(sum)
home_gb.aggregate(sum)
home_gb.agg({'FTHG':sum, 'FTAG':sum})

def ftr(x):
    return x.value_counts()['H']

home_gb.agg({'FTHG' : sum,'FTAG' : sum,'FTR' : ftr})

home_goal_df = home_gb.agg({'FTHG' : sum,'FTAG' : sum})
```

```python
def ftr2(x):
    return x.value_counts()

home_ftr_df = pd.DataFrame(home_gb['FTR'].apply(ftr2).unstack())

home_ftr_df.rename(columns={'A': 'Lost','D': 'Draw' , 'H': 'Win'}, inplace=True)

#홈에서 한 번도 진적이 없는 NaN이 뜨는 첼시를 위한 특별 처리
#DataFrame의 모든 NaN값을 0으로 채운다. 

home_ftr_df = home_ftr_df.fillna(0)
home_ftr_df.head()
```

```python
away_gb = working_df.groupby('AwayTeam')
away_goal_df = away_gb.agg({'FTHG' : sum, 'FTAG' : sum })

away_ftr_df = pd.DataFrame(away_gb['FTR'].apply(ftr2).unstack())

#Away 팀의 관점이므로 A는 승리, H는 패배
away_ftr_df.rename(columns={'A': 'Win','D': 'Draw' , 'H': 'Lost'}, inplace=True)

#away_ftr_df
#home_ftr_df

# 홈과 원정의 값들을 서로 더합니다. 
final_ftr = away_ftr_df + home_ftr_df

# 보기싫은 인덱스 이름 제거
#final_ftr.index.name = None

final_ftr

#승점 계산
final_ftr['Point']  = final_ftr['Win'] * 3 + final_ftr['Draw']

#승점 내림 차순으로 정렬
final_ftr.sort_values('Point', ascending = False)
```

```python
home_goal_df.rename(columns={'FTAG': 'GA', 'FTHG' : 'GF'}, inplace=True)
away_goal_df.rename(columns={'FTAG': 'GF', 'FTHG' : 'GA'}, inplace=True)

#홈 /원정 경기에 대한 골 득실 값을 모두 더해 줍니다. 

final_goal = home_goal_df + away_goal_df

# 골득실 계산

final_goal['GD'] = final_goal['GF']  - final_goal['GA']
final_goal.index.name = None
final_goal
final_goal.sort_values('GD', ascending=False)
```

```python
mission_completed = pd.concat([final_ftr,final_goal],axis=1)
mission_completed = mission_completed.reindex_axis(['Win','Draw','Lost','GF','GA','GD','Point'], axis=1)

#승점 우선 정렬, 동률인 경우 골득실로 정렬
mission_completed.sort_values(['Point','GD'], ascending = False)
```
